```python
import pandas as pd

# Load the property dataset
property_df = pd.read_csv('/content/property (1).csv')

# Display the first 5 rows of the DataFrame
display(property_df.head())
```

```python
# Display column information to check data types and non-null counts
display(property_df.info())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Define independent variables (X) and dependent variable (y)
X_cols_model1 = ['Square_Footage', 'Num_Bedrooms', 'Num_Bathrooms', 'Year_Built', 'Lot_Size']
X_model1 = property_df[X_cols_model1]
y_model1 = property_df['Price']

# Split the data into training and testing sets (80% train, 20% test)
X_train_model1, X_test_model1, y_train_model1, y_test_model1 = train_test_split(
    X_model1, y_model1, test_size=0.2, random_state=42
)

# Initialize and train the Linear Regression model
model1 = LinearRegression()
model1.fit(X_train_model1, y_train_model1)

# Make predictions on the test set
y_pred_model1 = model1.predict(X_test_model1)

# Calculate the R-squared value
r_squared_model1 = r2_score(y_test_model1, y_pred_model1)

print(f"Model 1 (Features: {', '.join(X_cols_model1)}) R-squared: {r_squared_model1:.4f}")
```

```python
# Define independent variables (X) for Model 2 and dependent variable (y)
X_cols_model2 = ['Square_Footage', 'Num_Bedrooms', 'Num_Bathrooms', 'Lot_Size']
X_model2 = property_df[X_cols_model2]
y_model2 = property_df['Price']

# Split the data into training and testing sets (using the same split strategy as Model 1 for fair comparison)
X_train_model2, X_test_model2, y_train_model2, y_test_model2 = train_test_split(
    X_model2, y_model2, test_size=0.2, random_state=42
)

# Initialize and train the Linear Regression model
model2 = LinearRegression()
model2.fit(X_train_model2, y_train_model2)

# Make predictions on the test set
y_pred_model2 = model2.predict(X_test_model2)

# Calculate the R-squared value
r_squared_model2 = r2_score(y_test_model2, y_pred_model2)

print(f"Model 2 (Features: {', '.join(X_cols_model2)}) R-squared: {r_squared_model2:.4f}")
print(f"Model 1 R-squared: {r_squared_model1:.4f}")
```

```python
# Define independent variables (X) for Model 3 and dependent variable (y)
X_cols_model3 = ['Square_Footage', 'Num_Bathrooms', 'Year_Built']
X_model3 = property_df[X_cols_model3]
y_model3 = property_df['Price']

# Split the data into training and testing sets (using the same split strategy for fair comparison)
X_train_model3, X_test_model3, y_train_model3, y_test_model3 = train_test_split(
    X_model3, y_model3, test_size=0.2, random_state=42
)

# Initialize and train the Linear Regression model
model3 = LinearRegression()
model3.fit(X_train_model3, y_train_model3)

# Make predictions on the test set
y_pred_model3 = model3.predict(X_test_model3)

# Calculate the R-squared value
r_squared_model3 = r2_score(y_test_model3, y_pred_model3)

print(f"Model 3 (Features: {', '.join(X_cols_model3)}) R-squared: {r_squared_model3:.4f}")
print(f"Model 1 R-squared: {r_squared_model1:.4f}")
print(f"Model 2 R-squared: {r_squared_model2:.4f}")
```
